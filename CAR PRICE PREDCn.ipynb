```python
car_body = pd.get_dummies(auto['carbody'],drop_first = True)
auto = pd.concat([auto,car_body])
```

```python
engine_type = pd.get_dummies(auto['enginetype'],drop_first = True)
auto = pd.concat([auto,engine_type])
```

```python
cylin_der = pd.get_dummies(auto['cylindernumber'],drop_first = True)
auto = pd.concat([auto,cylin_der])
```

```python
fuel_system = pd.get_dummies(auto['fuelsystem'],drop_first = True)
auto = pd.concat([auto,fuel_system])
```
